# Uso de gráficos para mostrar la información relativa a las características del conjunto de datos.


## Tarifa vs Distancia

A continuación veremos un gráfico que nos muestra la relación entre las tarifas (fare) y las distancias (distance) que recorren los taxis:  

```python
sns.scatterplot(df, x="distance",y="fare", hue=df['color'])
plt.title('Tarifa vs Distancia')
plt.show()
```

Como podemos ver tenemos una relación lineal positiva muy clara, a mayor distancia recorren los taxis, mayor es la tarifa que se debe pagar. También curiosamente vemos algunos taxis un poco dispersos ¿podrán ser estos atípicos? puede que lo veamos más adelante.

## Propinas vs Tarifas

A continuación también hemos generado otro gráficco para visualizar la relación que tienen las propinas (tip) con respecto a las tarifas (fare):

```python
sns.scatterplot(df, x="fare",y="tip", hue=df['color'])
plt.title('Propina vs Tarifa')
plt.show()
```

Cómo podemos ver a mayor tarifa vemos una pequeña tendencia a que se de una mayor propina, y se ven algunos datos dispersos también.

## Tarifa vs Total


A continuación veremos otro gráfico para ver como influye la tarifa sobre el total del precio de cada viaje:

```python
sns.scatterplot(df, x="fare",y="total", hue=df['color'])
plt.title('Tarifa vs Total')
plt.show()
```

Como podemos ver es bastante fuerte la correlación de estas dos variables, podemos decir qeu el total está dominado por la tarifa base.

# Realización de un análisis de correlación entre las características y explicación de su posible correlación.


Para realizar un análisis de correlación nos vamos a basar en el siguiente mapa o matriz que nos mostrará la correlación de cada feature con otro y consigo mismo:

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f", annot_kws={"size": 10})
plt.show()
```

Como podemos ver tienen el máximo de correlación  entre cada variable consigo misma. Por otro lado, vemos que hay también correlaciones muy fuertes superando el 90% entre la distancia y la tarifa y estas 2 también bastante correlacionadas con el total del coste de los viajes, por lo que según la teoría tendríamos razones para eliminarlas del dataset.

# Estudio y detección de los valores atípicos del conjunto de datos.


Para hacer el estudio y detección de los valores atípicos, he intentado aplicar la metodología de la clase, pero me pareció que iba a tener problemas, ya que como habíamos detectado en un principio, los datos de la media con el 50% eran muy distintos y otras peculiaridades, por lo cual me he apoyado un poco en la IA para generar lo siguiente:

```python

# Código generado con IA debido a que no tuve la capacidad de realizar esta detección con el método explicado en la clase.

# El unico ajuste hecho por mi es quitar los pasajeros ya que no hemos hecho tanto foco en esta variable para abordar el presente AED

num_cols = taxis.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(14,10))
for i, col in enumerate(num_cols):
    if col == 'passengers':
      continue
    plt.subplot(3,3,i+1)
    sns.boxplot(data=taxis, y=col)
    plt.title(col)

plt.tight_layout()
plt.show()
```

Según los gráficos que tenemos, podemos ver bastantes valores atípicos, por ejemplo el caso más curioso, es el de tolls (peajes), vemos 0 o cerca de cero valores típicos, contrasta bastante con los valores atípicos, pero esto puede deberse a que en casos de largas distancias sea donde se den, por ejemplo si observamos el gráfico de distancias, también vemos bastantes valores atípicos en casos de distancia mayores a 25.

Cómo habíamos observado anteriormente, aquí también vemos un patrón similar, en los gráficos de fare (tarifa), distancias y total.
